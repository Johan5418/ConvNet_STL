### 1. Libraries Loading and Image Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import torch 
import tensorflow as tf

#from PIL import Image
import cv2
```

Note that: cv2 is using BGR instead of RGB color, so you might need to convert before preprocessing

```python
#img = Image.open('Camera_for_preprocessing.jpg')
img = cv2.imread('picture_samples/Camera_for_preprocessing.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #Convert the color
_ = plt.imshow(img)
```

### 2. Resizing the image

```python
new_size = (256, 256)
resized_img = cv2.resize(img, (256, 256), interpolation=cv2.INTER_AREA)
_ = plt.imshow(resized_img)
```

### 3. Color Scaling

```python
imggray = cv2.cvtColor(resized_img, cv2.COLOR_RGB2GRAY)
_ = plt.imshow(imggray)
```

### 4. Normalization / Brightness Constrast 
- Try Histogram Equalization, Normalisation

```python
normalized_img = imggray / 255.0
_ = plt.imshow(normalized_img)
```

```python
eq_img = cv2.equalizeHist(imggray)
_ = plt.imshow(eq_img)
```

```python
mean, std = cv2.meanStdDev(imggray)
std_img = (imggray - mean) / std
_ = plt.imshow(std_img)
```

### 5. Noise Reduction
- Try Gaussian Smoothing

```python
# Apply Gaussian Blur
# Parameters: image, kernel size (width, height), sigmaX
# blurred_image = cv2.GaussianBlur(your_image, (5, 5), 1.0)
```

```python
reduced_img = cv2.GaussianBlur(eq_img, (7,7), 1.0) #Input, Kernel Size (Width, Height), SigmaX
_ = plt.imshow(reduced_img)
```
